```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
plt.rcParams['figure.dpi'] = 300
font = {'family' : 'Arial',
        'weight' : 'normal',
        'size'   : 5}
plt.rc('font', **font)
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['lines.linewidth']=0.7
plt.rcParams['xtick.direction']='in'
plt.rcParams['ytick.direction']='in'
plt.rcParams['xtick.major.width']=0.4
plt.rcParams['ytick.major.width']=0.4
plt.rcParams['xtick.major.size']=2
plt.rcParams['ytick.major.size']=2
```

```python
fig,ax = plt.subplots(1,1,figsize=(7.0,7.0*4/16))
# ax.spines['right'].set_visible(False)
# ax.spines['top'].set_visible(False)

h1 = pd.read_csv('armyf.csv',names=['time','fixh'])
time1,xh = np.array(h1['time']),np.array(h1['fixh'])
h2 = pd.read_csv('army.csv', names=['time','floath'])
time2,oh = np.array(h2['time']),np.array(h2['floath'])

ax.plot(time1,xh,'b')
ax.plot(time2,oh,'r')

leg = ax.legend(["fix arm","controlled arm"],loc=1,ncol=2)
leg.get_frame().set_linewidth(0.1)
# ax.set_ylim([1.76,1.85])
ax.set_xlabel('Time (s)',fontsize=7)
ax.set_ylabel('End-effector Height (m)',fontsize=7)
# ax[0].xaxis.set_label_coords(.5,-0.15)
# ax[0].yaxis.set_label_coords(-.15,.5)
```
